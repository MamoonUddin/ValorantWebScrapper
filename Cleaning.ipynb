```python
#Data Clean going from smallest to large
EditedMapData = MapData.copy()
for col in EditedMapData.columns:
    if col not in ['User Name','Tag'] :
        EditedMapData[col] = EditedMapData[col].fillna(0).astype(int)

UserCol = EditedMapData['User Name'] + EditedMapData["Tag"]

EditedMapData.insert(0,'User',UserCol)
EditedMapData = EditedMapData.drop(columns= ['User Name', 'Tag'])

EditedMapData.insert(1,"Matches", EditedMapData[[i for i in EditedMapData.columns if i != 'User']].sum(axis=1))

EditedMapData
```

```python
EditedUserData = UserData.copy()
UserCol = EditedUserData['User Name'] + EditedUserData["Tag"]
EditedUserData.insert(0,'User',UserCol)
EditedUserData = EditedUserData.drop(columns= ['User Name', 'Tag'])

CurentCol = EditedUserData['currentRank'] + '-' + EditedUserData["currentMMR"]
EditedUserData.insert(1,'CurrentRank',CurentCol)
EditedUserData = EditedUserData.drop(columns= ['currentRank', 'currentMMR'])

PeakCol = EditedUserData['peakRank'] + '-' + EditedUserData["peakMMR"]
EditedUserData.insert(2,'PeakRank',PeakCol)
EditedUserData = EditedUserData.drop(columns= ['peakRank', 'peakMMR'])

matches = EditedUserData['Wins'] + EditedUserData["Loss"]
matches = matches.astype('int')
EditedUserData.insert(3,'Matches',matches)
EditedUserData = EditedUserData.drop(columns= ['Wins', 'Loss'])

hits = EditedUserData['bodyShotHit'] + EditedUserData["legShotHit"] + EditedUserData["headShotHit"]
hits = hits.astype('int')
EditedUserData.insert(3,'Total Hits',hits)
EditedUserData = EditedUserData.drop(columns= ['bodyShotHit', 'legShotHit', 'headShotHit' ])

EditedUserData
```
